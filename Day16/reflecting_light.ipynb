```python
import numpy as np

def beam_path(r,c,d,grid,energized):
    beam_stack = [(r,c,d)]
    dejavu = set()

    while beam_stack:
        i,j,dir = beam_stack.pop(0)

        if (i,j,dir) in dejavu:
            continue
        else:
            dejavu.add((i,j,dir))

        if j != -1:
            energized.add((i,j))
        
        match dir:
            case 'u':
                if i != 0:
                    next_step = (i-1,j)
                else:
                    continue
            case 'd':
                if i != grid.shape[0]-1:
                    next_step = (i+1,j)
                else:
                    continue
            case 'l':
                if j != 0:
                    next_step = (i,j-1)
                else:
                    continue
            case 'r':
                if j != grid.shape[1]-1:
                    next_step = (i,j+1)
                else:
                    continue


        match grid[next_step]:
            case '/':
                if dir == 'u':
                    beam_stack.append((*next_step,'r'))
                    continue
                elif dir == 'd':
                    beam_stack.append((*next_step,'l'))
                    continue
                elif dir == 'l':
                    beam_stack.append((*next_step,'d'))
                    continue
                else:
                    beam_stack.append((*next_step,'u'))
                    continue
            case "\\":
                if dir == 'u':
                    beam_stack.append((*next_step,'l'))
                    continue
                elif dir == 'd':
                    beam_stack.append((*next_step,'r'))
                    continue
                elif dir == 'l':
                    beam_stack.append((*next_step,'u'))
                    continue
                else:
                    beam_stack.append((*next_step,'d'))
                    continue
            case '|':
                if dir == 'u' or dir == 'd':
                    beam_stack.append((*next_step,dir))
                    continue
                else:
                    beam_stack.append((*next_step,'u'))
                    beam_stack.append((*next_step,'d'))
                    continue
            case '-':
                if dir == 'l' or dir == 'r':
                    beam_stack.append((*next_step,dir))
                    continue
                else:
                    beam_stack.append((*next_step,'l'))
                    beam_stack.append((*next_step,'r'))
                    continue
            case _:
                beam_stack.append((*next_step,dir))
                continue

    return
```

```python
grid = []
with open("input.txt") as f:
    for line in f:
        grid.append(list(line)[:-1])

grid = np.array(grid)
energized = set()

beam_path(0,-1,'r',grid, energized)

int(len(energized))
```

```python
edges = [(i,j) for i in [0,grid.shape[0]-1] for j in range(grid.shape[1])] + \
        [(i,j) for i in range(grid.shape[0]) for j in [0,grid.shape[1]-1]]

res = []
for edge in edges:
    energized = set()
    if edge == (0,0):
        beam_path(0,-1,'r',grid,energized)
        res.append(len(energized))
        energized = set()
        beam_path(-1,0,'d',grid,energized)
        res.append(len(energized))
    
    elif edge == (0,109):
        beam_path(0,110,'l',grid,energized)
        res.append(len(energized))
        energized = set()
        beam_path(-1,109,'d',grid,energized)
        res.append(len(energized))

    elif edge == (109,0):
        beam_path(109,-1,'r',grid,energized)
        res.append(len(energized))
        energized = set()
        beam_path(110,0,'u',grid,energized)
        res.append(len(energized))

    elif edge == (109,109):
        beam_path(109,110,'l',grid,energized)
        res.append(len(energized))
        energized = set()
        beam_path(110,109,'u',grid,energized)
        res.append(len(energized))

    else:
        if edge[0] == 0:
            beam_path(*edge,'d',grid,energized)
            res.append(len(energized))
        elif edge[1] == 0:
            beam_path(*edge,'r',grid,energized)
            res.append(len(energized))
        elif edge[0] == 109:
            beam_path(*edge,'u',grid,energized)
            res.append(len(energized))
        else:
            beam_path(*edge,'l',grid,energized)
            res.append(len(energized))

max(res)           
```
